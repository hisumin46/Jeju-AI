# Transfer Learning (전이학습)

```python
import numpy as np
import cv2
from tensorflow.keras import datasets
from tensorflow.keras import applications

from tensorflow.keras import models, layers
```

```python
def changeData(data,size=(32,32)):
    ans = []
    for i in range(len(data)):
        img = data[i]
        cv2.imwrite("t.png",img)
        img = cv2.imread("t.png")
        img = cv2.resize(img,size)
        ans.append(img)
    ans = np.array(ans)
    return ans
        
```

```python
def getData():
    (x_train,y_train),(x_test,y_test) = datasets.mnist.load_data()
    print(x_train.shape,y_train.shape)
    x_train = changeData(x_train)
    x_test = changeData(x_test)
    
    return (x_train,y_train),(x_test,y_test)
getData()
```

```python


def makeModel():
    app_model = applications.VGG16(include_top=False, input_shape=(32,32,3))
    app_model.trainable = False
    
    use_model = models.Sequential()
    use_model.add(app_model)
    
    use_model.add(layers.Flatten())
    use_model.add(layers.Dense(256,activation = 'relu'))
    use_model.add(layers.Dense(10,activation = 'softmax'))
    use_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
    print(use_model.summary())
    return use_model
    
makeModel()
    

def doLearn():
    (x_train,y_train),(x_test,y_test) = getData() 
    model = makeModel()
    
    log = model.fit(x_train,y_train,epochs=2,batch_size=16)
    score = model.evaluate(x_train,y_train)
    print(score)

doLearn()
```
