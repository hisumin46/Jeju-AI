# Transfer Learning (전이학습)

```python
import numpy as np
import cv2
from tensorflow.keras import datasets
from tensorflow.keras import applications

from tensorflow.keras import models, layers
```

```python
def changeData(data,size=(32,32)):
    ans = []
    for i in range(len(data)):
        img = data[i]
        cv2.imwrite("t.png",img)
        img = cv2.imread("t.png")
        img = cv2.resize(img,size)
        ans.append(img)
    ans = np.array(ans)
    return ans
        
```

```python
def getData():
    (x_train,y_train),(x_test,y_test) = datasets.mnist.load_data()
    print(x_train.shape,y_train.shape)
    x_train = changeData(x_train)
    x_test = changeData(x_test)
    
    return (x_train,y_train),(x_test,y_test)
getData()
```

```python


def makeModel():
    app_model = applications.VGG16(include_top=False, input_shape=(32,32,3))
    app_model.trainable = False
    
    use_model = models.Sequential()
    use_model.add(app_model)
    
    use_model.add(layers.Flatten())
    use_model.add(layers.Dense(256,activation = 'relu'))
    use_model.add(layers.Dense(10,activation = 'softmax'))
    use_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
    print(use_model.summary())
    return use_model
    
makeModel()
    

def doLearn():
    (x_train,y_train),(x_test,y_test) = getData() 
    model = makeModel()
    
    log = model.fit(x_train,y_train,epochs=2,batch_size=16)
    score = model.evaluate(x_test,y_test)
    print(score)

doLearn()
```

```python
import numpy as np
import cv2
import pandas as pd
import os
from tensorflow.keras import applications,models,layers
```

```python
print(os.listdir("./cub200_5/train/001.Black_footed_Albatross/")[0])

```

```python
def get_XY(path):
    x = []
    y = []
    
    folder_list = os.listdir(path)
    print(folder_list)
    
    for i in range(len(folder_list)):
        file_path = path + "/" + folder_list[i]
        file_name = os.listdir(file_path)
        
        for j in range(len(file_name)):
            full_file_path = file_path +"/" + file_name[j]
            #print(full_file_path)
            img = cv2.imread(full_file_path)
            img = cv2.resize(img,(32,32))
            x.append(img)
            y.append(i)
    x = np.array(x)
    y = np.array(y)
    print("x_shape : ",x.shape)
    print("y_shape : ",y.shape)
    return x,y
        
```

```python
def getData2():
    path = "./cub200_5/train"
    x_train,y_train = get_XY(path)
    
    
    path = "./cub200_5/test"
    x_test,y_test = get_XY(path)
    
    return (x_train,y_train),(x_test,y_test)


```

```python
def makeModel2():
    app_model = applications.resnet50.ResNet50(
        include_top = False,
        weights = "imagenet",
        input_shape = (32,32,3)
    )
    app_model.trainable = False
    
    use_model = models.Sequential()
    use_model.add(app_model)
    use_model.add(layers.Flatten())
    use_model.add(layers.Dense(1024,activation='relu'))
    use_model.add(layers.Dense(5,activation="softmax"))
    use_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
    print(use_model.summary())
    
    return use_model
makeModel2()
    
```

```python
def doLearn():
    (x_train,y_train),(x_test,y_test) = getData2()
    model = makeModel2()
    
    log = model.fit(x_train,y_train,epochs=10,batch_size=16)
    score = model.evaluate(x_test,y_test)
    print(score)
    
doLearn()
```
