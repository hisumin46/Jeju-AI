# Transfer Learning (전이학습)

```python
import numpy as np
import cv2
from tensorflow.keras import datasets
from tensorflow.keras import applications

from tensorflow.keras import models, layers
```

```python
def changeData(data,size=(32,32)):
    ans = []
    for i in range(len(data)):
        img = data[i]
        cv2.imwrite("t.png",img)
        img = cv2.imread("t.png")
        img = cv2.resize(img,size)
        ans.append(img)
    ans = np.array(ans)
    return ans
        
```

```python
def getData():
    (x_train,y_train),(x_test,y_test) = datasets.mnist.load_data()
    print(x_train.shape,y_train.shape)
    x_train = changeData(x_train)
    x_test = changeData(x_test)
    
    return (x_train,y_train),(x_test,y_test)
getData()
```

```python


def makeModel():
    app_model = applications.VGG16(include_top=False, input_shape=(32,32,3))
    app_model.trainable = False
    
    use_model = models.Sequential()
    use_model.add(app_model)
    
    use_model.add(layers.Flatten())
    use_model.add(layers.Dense(256,activation = 'relu'))
    use_model.add(layers.Dense(10,activation = 'softmax'))
    use_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
    print(use_model.summary())
    return use_model
    
makeModel()
    

def doLearn():
    (x_train,y_train),(x_test,y_test) = getData() 
    model = makeModel()
    
    log = model.fit(x_train,y_train,epochs=2,batch_size=16)
    score = model.evaluate(x_test,y_test)
    print(score)

doLearn()
```

```python
import numpy as np
import cv2
import pandas as pd
import os
from tensorflow.keras import applications,models,layers
```

```python
print(os.listdir("./cub200_5/train/001.Black_footed_Albatross/")[0])

```

```python
def get_XY(path):
    x = []
    y = []
    
    folder_list = os.listdir(path)
    print(folder_list)
    
    for i in range(len(folder_list)):
        file_path = path + "/" + folder_list[i]
        file_name = os.listdir(file_path)
        
        for j in range(len(file_name)):
            full_file_path = file_path +"/" + file_name[j]
            #print(full_file_path)
            img = cv2.imread(full_file_path)
            img = cv2.resize(img,(32,32))
            x.append(img)
            y.append(i)
    x = np.array(x)
    y = np.array(y)
    print("x_shape : ",x.shape)
    print("y_shape : ",y.shape)
    return x,y
        
```

```python
def getData2():
    path = "./cub200_5/train"
    x_train,y_train = get_XY(path)
    
    
    path = "./cub200_5/test"
    x_test,y_test = get_XY(path)
    
    return (x_train,y_train),(x_test,y_test)


```

```python
def makeModel2():
    app_model = applications.resnet50.ResNet50(
        include_top = False,
        weights = "imagenet",
        input_shape = (32,32,3)
    )
    app_model.trainable = False
    
    use_model = models.Sequential()
    use_model.add(app_model)
    use_model.add(layers.Flatten())
    use_model.add(layers.Dense(1024,activation='relu'))
    use_model.add(layers.Dense(5,activation="softmax"))
    use_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
    print(use_model.summary())
    
    return use_model
makeModel2()
    
```

```python
def doLearn():
    (x_train,y_train),(x_test,y_test) = getData2()
    model = makeModel2()
    
    log = model.fit(x_train,y_train,epochs=10,batch_size=16)
    score = model.evaluate(x_test,y_test)
    print(score)
    
doLearn()
```

```python
from djitellopy import Tello
from threading import Thread
import time
import cv2
```

```python
def streamTello(drone):
    drone.connect()
    drone.streamon()
    time.sleep(3) #3초 대기를 주지 않으면 오류가 발생할 수 있음

    while True:
        img = drone.get_frame_read().frame
        img = cv2.resize(img, (340,240))
        cv2.imshow("340*240", img)
        Key = cv2.waitKey(1)
        if Key == ord("q"):
            break

    cv2.destroyAllWindows()
    drone.streamoff()
    drone.end()
```

```python
def flyTello(drone):
    drone.connect()
    power = drone.get_battery()

    if power<30:
        print("에너지가 부족함")
    else:
        while True:
            if power<30:
                print("에너지가 부족함")
                break
            print("이륙 착륙 앞쪽 뒷쪽 우측 좌측 시계 반시계")
            print("[5]  [.]  [8]  [2]  [6]  [4] [86] [42] ")
            cmd = input("명령어를 넣어주세요 >> ")
            
            if cmd=="end":
                print("프로그램을 종료하겠습니다.")
                break
            elif cmd=="up":
                drone.takeoff() 
            elif cmd=="down":
                drone.land() #내려오기
            elif cmd=="go":
                drone.move_forward(20) #앞으로 이동
            elif cmd=="back":
                drone.move_back(20) #뒤로 이동
            elif cmd=="left":
                drone.move_left(20) #왼쪽으로 이동
            elif cmd=="right":
                drone.move_right(20) #오른쪽으로 이동
            elif cmd=="clock_rotate":
                drone.rotate_clockwise(90) #시계방향
            elif cmd=="reverse_clock":
                drone.rotate_counter_clockwise(90) #반시계방향
            else:
                print("알수없는 명령어!")
```

```python
def main():
    drone = Tello()
    videoRun = Thread(target=streamTello, args=[drone])
    
    flyRun = Thread(target=flyTello, args=[drone])

    videoRun.start()
    time.sleep(10)
    flyRun.start()
    time.sleep(1)

    videoRun.join()
    flyRun.join()

    drone.land()
    time.sleep(3)
    drone.end()

main()
```
