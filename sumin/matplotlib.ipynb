### 두개 축 만들기 - twinx()

```python
data1 = np.random.randint(0,10,(20,))
data2 = np.random.rand(20,)
plt.plot(data1)
plt.plot(data2)
plt.show()
```

```python
# 축을 먼저 가져와서 
y1 = plt.axes()
y1.plot(data1, color="r")

# 두번째 축을 가져옴
y2 = y1.twinx()
y2.plot(data2)
plt.show()
```

```python
data1 = np.random.randint(0,100,(30,3))
data2 = np.random.rand(30,2) # 2개의 배열로 으로 30개

y1 = plt.axes()
y1.plot(data1[:,0], color="r", label="nono")
y1.legend(loc = "upper left")
y2 = plt.twinx()
y2.plot(data2[:,0], color="b", label="goob")
y2.legend(loc = "upper right")
plt.grid()
plt.show()
```

### 과제 - 7과목 성적 데이터 분석

```python
from turtle import color
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def avg_student():
  # 랜덤 데이터 생성
  data = np.random.randint(0,100,(200,7))

  # 과목평균 리스트 초기화
  avg_list = []


  # 데이터 프레임 만들기
  df = pd.DataFrame(data,columns=["subject_1","subject_2","subject_3","subject_4","subject_5","subject_6","subject_7"])

  # 과목 평균    
  for i in range(0,7):
    subject = data[:,i]
    subject = sum(subject)//200
    avg_list.append(subject)

  # 학생별 평균
  for i in range(200):
      df.loc[i,"tot"] = sum(data[i,:])
      df.loc[i,"avg"] = df.loc[i,"tot"]//7

  # 데이터 분석 - 시각화
  # 과목별 평균 바플롯
  plt.figure(figsize=(8,8))
  plt.bar(np.arange(7), avg_list)
  plt.title("avg_subject")
  plt.xlabel("subject")
  plt.ylabel("score")
  plt.xticks(range(0,len(df.columns[:7])), df.columns[:7])
  plt.show()

  # 과목별 평균 점수 박스 플롯
  plt.figure(figsize=(8,6))
  plt.boxplot([data[:,0], data[:,1],data[:,2], data[:,3], data[:,4], data[:,5], data[:,6]], labels=["subject_1","subject_2","subject_3","subject_4","subject_5","subject_6","subject_7"])
  plt.grid()
  plt.show()

  
avg_student()
```
