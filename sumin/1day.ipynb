```python
import random

def getcom():
  a = [ ]
  for b in range(3):
    b = random.randint(10,20)
    while b in a:
      b = random.randint(10,20)
    a.append(b)


def getplayer():
  a = []
  for b in range(3):
    b = int(input("0~9 사이 값을 입력:"))
    while b in a:
      b = int(input("중복입니다 0~9 사이 값을 입력:"))
    a.append(b)



def getsb(com, player):
  while True:
    strike = 0
    ball = 0


            
    list_a = [com,player]
    for i in range(len(list_a[0])):
        if list_a[0][i] in list_a[1]:
            flag = 0
            for j in range(len(list_a[1])):
                if list_a[0][i] == list_a[1][j] and i==j:
                    strike+=1
                    flag = 1
                    break
            if flag ==0:
                ball+=1
    print("strike :",strike,"ball :", ball)
    if strike == 3:
        print("정답입니다")
        break



def game():
  com = getcom()
  for i in range(5):
    player = getplayer()
    s,b = getsb(com, player)
    print(f"스트라이크 : {s}  볼 : {b}")
    
    if (s == 3):
      print("이김", com)
      return
    print("짐", com)

game()
```

```python
from turtle import color
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def avg_student():
  # 랜덤 데이터 생성
  data = np.random.randint(0,100,(200,7))

  # 과목평균 리스트 초기화
  avg_list = []


  # 데이터 프레임 만들기
  df = pd.DataFrame(data,columns=["subject_1","subject_2","subject_3","subject_4","subject_5","subject_6","subject_7"])

  # 과목 평균    
  for i in range(0,7):
    subject = data[:,i]
    subject = sum(subject)//200
    avg_list.append(subject)

  # 학생별 평균
  for i in range(200):
      df.loc[i,"tot"] = sum(data[i,:])
      df.loc[i,"avg"] = df.loc[i,"tot"]//7

  # 데이터 분석 - 시각화
  # 과목별 평균 바플롯
  plt.figure(figsize=(8,8))
  plt.bar(np.arange(7), avg_list)
  plt.title("avg_subject")
  plt.xlabel("subject")
  plt.ylabel("score")
  plt.xticks(range(0,len(df.columns[:7])), df.columns[:7])
  plt.show()

  # 과목별 평균 점수 박스 플롯
  plt.figure(figsize=(8,6))
  plt.boxplot([data[:,0], data[:,1],data[:,2], data[:,3], data[:,4], data[:,5], data[:,6]], labels=["subject_1","subject_2","subject_3","subject_4","subject_5","subject_6","subject_7"])
  plt.grid()
  plt.show()

  
avg_student()
```
