### 파이썬 기초

```python
import random

def getcom():
  a = [ ]
  for b in range(3):
    b = random.randint(10,20)
    while b in a:
      b = random.randint(10,20)
    a.append(b)


def getplayer():
  a = []
  for b in range(3):
    b = int(input("0~9 사이 값을 입력:"))
    while b in a:
      b = int(input("중복입니다 0~9 사이 값을 입력:"))
    a.append(b)



def getsb(com, player):
  while True:
    strike = 0
    ball = 0


            
    list_a = [com,player]
    for i in range(len(list_a[0])):
        if list_a[0][i] in list_a[1]:
            flag = 0
            for j in range(len(list_a[1])):
                if list_a[0][i] == list_a[1][j] and i==j:
                    strike+=1
                    flag = 1
                    break
            if flag ==0:
                ball+=1
    print("strike :",strike,"ball :", ball)
    if strike == 3:
        print("정답입니다")
        break



def game():
  com = getcom()
  for i in range(5):
    player = getplayer()
    s,b = getsb(com, player)
    print(f"스트라이크 : {s}  볼 : {b}")
    
    if (s == 3):
      print("이김", com)
      return
    print("짐", com)

game()
```
